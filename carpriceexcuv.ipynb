```python
numeric_cols = df.select_dtypes(include=np.number).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
```

```python
print(df.columns)
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
print(df.columns)
```

```python
df["engine_volume"] = (
    df["engine_volume"]
    .astype(str)
    .str.replace("Turbo", "", regex=False)
    .str.strip()
)

df["engine_volume"] = pd.to_numeric(df["engine_volume"], errors="coerce")

df["engine_volume"] = df["engine_volume"].fillna(df["engine_volume"].mean())
```

```python
df["airbags"] = df["airbags"].fillna(df["airbags"].mean())
```

```python
df.isnull().sum()
```

```python
def cap_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[column] = np.where(df[column] < lower, lower,
                           np.where(df[column] > upper, upper, df[column]))
```

```python
cap_outliers(df, "price")
cap_outliers(df, "kilometers_driven")
cap_outliers(df, "engine_volume")
```
